```python
df = pd.read_sql_query("SELECT NOME_PROGRAMA_GOVERNO, COUNT(DISTINCT NOME_ORGAO) AS NUMERO_ORGAOS_DISTINTOS FROM ORGAO NATURAL JOIN DESPESA_PUBLICA NATURAL JOIN PROGRAMA_GOVERNO WHERE VALOR_DESPESA > 0 GROUP BY NOME_PROGRAMA_GOVERNO", conn)
print("Consulta 7")
df
```
