```python
df = pd.read_sql_query("SELECT COD_FUNCAO, NOME_FUNCAO FROM FUNCAO NATURAL JOIN DETALHAMENTO_FUNCAO NATURAL JOIN SUBFUNCAO WHERE NOME_SUBFUNCAO = (SELECT NOME_SUBFUNCAO FROM FUNCAO NATURAL JOIN DETALHAMENTO_FUNCAO NATURAL JOIN SUBFUNCAO GROUP BY NOME_SUBFUNCAO ORDER BY COUNT(NOME_SUBFUNCAO) DESC LIMIT 1)", conn)
print('Consulta 1')
df
```

```python
df = pd.read_sql_query("SELECT NOME_FUNCAO, NOME_SUBFUNCAO, SUM(VALOR_DESPESA) AS VALOR_TOTAL FROM FUNCAO NATURAL JOIN DETALHAMENTO_FUNCAO NATURAL JOIN SUBFUNCAO NATURAL JOIN DESPESA_PUBLICA WHERE ANO_DESPESA = 2018 AND MES_DESPESA > 7 GROUP BY NOME_FUNCAO, NOME_SUBFUNCAO ORDER BY VALOR_TOTAL DESC", conn)
print("Consulta 6")
df
```

```python
df = pd.read_sql_query("SELECT NOME_PROGRAMA_GOVERNO, COUNT(DISTINCT NOME_ORGAO) AS NUMERO_ORGAOS_DISTINTOS FROM ORGAO NATURAL JOIN DESPESA_PUBLICA NATURAL JOIN PROGRAMA_GOVERNO WHERE VALOR_DESPESA > 0 GROUP BY NOME_PROGRAMA_GOVERNO", conn)
print("Consulta 7")
df
```

```python
df = pd.read_sql_query("SELECT NOME_FUNCAO, NOME_SUBFUNCAO, AVG(VALOR_DESPESA) AS MEDIA_DESPESA_FUNCAO_SUBFUNCAO FROM FUNCAO NATURAL JOIN DETALHAMENTO_FUNCAO NATURAL JOIN SUBFUNCAO NATURAL JOIN DESPESA_PUBLICA GROUP BY NOME_FUNCAO, NOME_SUBFUNCAO HAVING MEDIA_DESPESA_FUNCAO_SUBFUNCAO > 1000", conn)
print("Consulta 2")
df
```

```python
df = pd.read_sql_query("SELECT DISTINCT NOME_PROGRAMA_GOVERNO, VALOR_DESPESA FROM DESPESA_PUBLICA NATURAL JOIN ORGAO NATURAL JOIN PROGRAMA_GOVERNO WHERE ANO_DESPESA = 2018 ORDER BY VALOR_DESPESA DESC", conn)
print("Consulta 3")
df
```

```python
df = pd.read_sql_query("SELECT DISTINCT MES_DESPESA, ANO_DESPESA, MODALIDADE_DESPESA, NOME_PROGRAMA_GOVERNO, VALOR_DESPESA FROM DESPESA_PUBLICA NATURAL JOIN ORGAO NATURAL JOIN PROGRAMA_GOVERNO WHERE ANO_DESPESA = 2018 AND NOME_ORGAO = 'Ministério da Educação' AND MES_DESPESA < 7 ORDER BY VALOR_DESPESA DESC", conn)
print("Consulta 4")
df
```

```python
df = pd.read_sql_query("SELECT DISTINCT NOME_FUNCAO, VALOR_DESPESA FROM FUNCAO NATURAL JOIN DETALHAMENTO_FUNCAO NATURAL JOIN DESPESA_PUBLICA", conn)
print("Consulta 12")
df
```

```python
df = pd.read_sql_query("SELECT DISTINCT MODALIDADE_DESPESA, VALOR_DESPESA, NOME_ORGAO FROM ORGAO NATURAL JOIN DESPESA_PUBLICA WHERE MODALIDADE_DESPESA = 'Reserva de Contingência'", conn)
print("Consulta 13")
df
```

```python
df = pd.read_sql_query("SELECT DISTINCT VALOR_DESPESA, NOME_PROGRAMA_GOVERNO FROM DESPESA_PUBLICA NATURAL JOIN PROGRAMA_GOVERNO WHERE NOME_PROGRAMA_GOVERNO = 'ENEM'", conn)
print("Consulta 14")
df
```
